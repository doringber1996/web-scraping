```python
def get_price(price_tag):
    try:
        price= price_tag.get_text()
        price_match = re.search(r'₪([\d,]+)', price)
        price = price_match.group(1)
        price = int(price.replace(',', ''))
    except:
        return None
    return price
```

```python
def get_rooms(room_tag):
    try:
        s=room_tag.get_text()
        number = re.findall(r'[\d.]+', s)[0]        
        
    except:
        return None
    return number
```

```python
def get_size(size_tag):
    try:
        s=size_tag.get_text()
        number = re.findall(r'[\d.]+', s)[0]
    except:
        return None
    return number
```

```python
def get_type(class_tag):
    try:
        s=class_tag.get_text()
    except:
        return None
    return s
```

```python
def adress_info(adress_tag): 
    try:
        details=adress_tag.get_text().split(',')
        area = details[1]
        street=details[0]
    except:
        area = None
        number=None
        street=None
        return street,number,area
    try:
        number = re.findall(r'\d+', street)[0] 
        street = re.search(r'(.+?)\d', street).group(1)
    except:
        number=None
        return street,number,area
    return street,number,area
```

```python
def get_info_from_txt(soup="dimona.txt"):
    import requests
    from bs4 import BeautifulSoup
    import pandas as pd
    import numpy as np
    import re
    import time
    from datetime import datetime
    property_price=np.array([])
    property_rooms=np.array([])
    property_size=np.array([])
    property_class=np.array([])
    property_street=np.array([])
    property_number_in_street=np.array([])
    property_city_area=np.array([])
    property_city = np.array([])
    try:
        with open(soup, 'r', encoding='utf-8') as file:
            html_content = file.read()
            S = BeautifulSoup(html_content,'html.parser')
        apartments_detailes=S.findAll('div',class_="css-l57rop eu6esji2",attrs={'data-auto': True})
        for i in apartments_detailes:
            if(i.find(class_="css-v4h09j e1sx3tzs16") == None):#מסנן מודעות פירסום שקיימות בדף
                continue
            else:
                property_price=np.append(property_price,get_price(i.find(attrs={'data-auto': "property-price"})))
                property_rooms=np.append(property_rooms,get_rooms(i.find(attrs={'data-auto': "property-rooms"})))
                property_size=np.append(property_size,get_size(i.find(attrs={'data-auto': "property-size"})))
                property_class=np.append(property_class,get_type(i.find(attrs={'data-auto': "property-class"})))
                property_street=np.append(property_street,adress_info(i.find(attrs={'data-auto': "property-address"}))[0])
                property_number_in_street=np.append(property_number_in_street,adress_info(i.find(attrs={'data-auto': "property-address"}))[1])
                property_city_area=np.append(property_city_area,adress_info(i.find(attrs={'data-auto': "property-address"}))[2])
                property_city=np.append(property_city,"דימונה")
                
    except:
        return None
    
    property_img=np.array([])
    property_floor_of=np.array([])
    property_hasElevator = np.array([])
    property_hasParking = np.array([])
    property_hasBars = np.array([])
    property_hasStorage =np.array([])
    property_condition = np.array([])
    property_hasAirCondition = np.array([])
    property_hasBalcony =np.array([])
    property_hasMamad = np.array([])
    property_handicapFriendly = np.array([])
    property_entranceDate = np.array([])
    property_furniture = np.array([])
    property_publishedDays = np.array([])
    property_description= np.array([])
    appartments_lst=[]
    for i in range(0,len(property_price)):
        with open("ap_"+str(i)+".txt", 'r', encoding='utf-8') as file:
            html_content = file.read()
            html_content=BeautifulSoup(html_content,'html.parser')
        appartments_lst.append(html_content)
    for i in appartments_lst:
            property_img=np.append(property_img,get_num_img(i))
            property_floor_of=np.append(property_floor_of,get_floor_out_of(i))
            property_hasElevator=np.append(property_hasElevator,get_elevator(i))
            property_hasParking=np.append(property_hasParking,get_parking(i))
            property_hasBars=np.append(property_hasBars,get_bars(i))
            property_hasStorage=np.append(property_hasStorage,get_storage(i))
            property_hasAirCondition=np.append(property_hasAirCondition,get_ac(i))
            property_hasBalcony=np.append(property_hasBalcony,get_balcony(i))
            property_hasMamad=np.append(property_hasMamad,get_mamad(i))
            property_handicapFriendly=np.append(property_handicapFriendly,get_handicap_friendly(i))
            property_condition=np.append(property_condition,get_condition(i))
            property_entranceDate=np.append(property_entranceDate,get_entrance_date(i))
            property_furniture=np.append(property_furniture,get_furniture(i))
            property_publishedDays=np.append(property_publishedDays,get_published_days(i))
            property_description=np.append(property_description,get_description(i))

    d = {'City':property_city,'type ':property_class ,'room_number':property_rooms , 'Area': property_size,'Street ':property_street,'number_in_street':property_number_in_street,'city_area':property_city_area,'price':property_price,'num_of_images':property_img,'floor_out_of':property_floor_of,'hasElevator ':property_hasElevator,'hasParking':property_hasParking,'hasBars':property_hasBars,'hasStorage':property_hasStorage,'hasAirCondition':property_hasAirCondition,'hasBalcony':property_hasBalcony,'hasMamad':property_hasMamad,'handicapFriendly':property_handicapFriendly,'condition':property_condition,'entranceDate':property_entranceDate,'furniture':property_furniture,'publishedDays':property_publishedDays,'description':property_description}
    df = pd.DataFrame(data=d)
    return df
```
